```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("NASDAQ100.csv", parse_dates=['Date'], index_col='Date')
print(df.head())
print(df.tail())
```

```python
#Dataset begins 17th of February 2015 (Tuesday), ends 14th of February 2025 (Friday). This is exactly 3650 days. The NASDAQ is closed on weekends, which is 1040 days, so 2610 days remaining. The remaining 94 days could be explained by holidays.
print(df.isna().sum())
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Close Price', color='blue')
plt.title('NASDAQ-100 Index close price')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
df.to_csv('NASDAQ-100_cleaned.csv', index=False)
```
